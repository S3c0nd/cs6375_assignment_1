```python
import json

f = open('training.json') 
f2 = open('validation.json') 
f3 = open('test.json')

train_set = json.load(f)
va_set = json.load(f2)
test_set = json.load(f3)
print('train sample numbers: ', len(train_set))
print('val sample numbers: ', len(va_set))
print('test sample numbers: ', len(test_set))

train_s = {}
val_s = {}
test_s = {}

for i in train_set:
    if str(i['stars']) not in train_s.keys():
        train_s[str(i['stars'])] = 1
    else:
        train_s[str(i['stars'])] += 1

for i in va_set:
    if str(i['stars']) not in val_s.keys():
        val_s[str(i['stars'])] = 1
    else:
        val_s[str(i['stars'])] += 1

for i in test_set:
    if str(i['stars']) not in test_s.keys():
        test_s[str(i['stars'])] = 1
    else:
        test_s[str(i['stars'])] += 1

print(train_s, val_s, test_s)

import matplotlib.pyplot as plt

# 示例字典


# 提取字典的键和值
keys = list(test_s.keys())
values = list(test_s.values())

# 创建柱状图
plt.bar(keys, values)

# 添加标题和标签
plt.title('Target distribution in test set')
plt.xlabel('stars')
plt.ylabel('numbers')

# 显示图形
plt.show()

```
